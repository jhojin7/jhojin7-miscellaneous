````python
arr = questions['question'].head(3)
import re
exp = '(https:\/\/.*\/)'
pattern = re.compile(exp)
# re.findall(pattern,arr[0])[0]

questions['question'] = [re.findall(pattern,questions['question'][x])[0] for x in range(len(questions))]
questions['slug'] = "slug"
questions['completed'] = True

questions
````

````python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
nums = [1,2,3,4,5]
head = ListNode()
p = head
for n in nums:
    p.val = n
    p.next = ListNode()
    print(p.val)
    p = p.next

p = head
while p != None:
    print("x",p.val)
    p = p.next

# #####################
# node, prev = head, None
# while node:
#     next, node.next = node.next, prev
#     prev, node = node, next
# #####################
# p = prev
# while p != None:
#     print("x",p.val)
#     p = p.next
````
