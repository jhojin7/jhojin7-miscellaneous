```python
import pandas as pd

f = open('algo-interview-questions.csv','r',encoding='utf8')
f_text = f.read()
# for md->csv, goto https://tableconvert.com/markdown-to-csv
questions = pd.read_csv("./algo-interview-questions.csv",sep='|')

questions = questions[['number','question','chapter','solutions']]
questions = questions.set_index('number')
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html

questions.head()
```
